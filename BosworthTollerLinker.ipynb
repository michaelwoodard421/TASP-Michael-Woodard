```python
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents

word = input()
print('\n')#take in word(s) as input  #will have to do string handling later for security/bugs 
BTurl = "https://bosworthtoller.com/search?q="
response=requests.get(BTurl + word)
soup = BeautifulSoup(response.text, 'html.parser')
entryTable = soup.find_all('div', class_ = "btd--search-entry btd--trailer-lm")

for i in range (0, len(entryTable)):
    header = entryTable[i].find('header')
    head = header.find('h3')
    type_ =header.find('div', class_="btd--search-entry-wordclass btd--margin-left-xss")
    for span in header.find_all("span", class_='base'): 
        span.decompose()
    plural_gender = header.find('section', class_="btd--entry-grammar btd--old-english")
 
    top_line = head.get_text()
    if type_:
        top_line = top_line + type_.get_text()
    if plural_gender:
        top_line = top_line + plural_gender.get_text()
    
    top_line_nice =' '.join(top_line.split())
    
    print(top_line_nice)
   
    body = entryTable[i].find('div', class_="btd--margin-no btd--search-description btd--trailer-xs")
    if body:
        subbodylist = body.find_all('div')
    
        i = 0
        while i < len(subbodylist):
            subbody = ' '.join(subbodylist[i].get_text().split())
            #print(i)
            print(subbody)
            i = i + 1
          
    linked_entries = entryTable[i].find('div', class_="btd--search-linked-entries btd--leader-xss btd--trailer-xss")
    if (linked_entries):
        linked_entriesF = ' '.join(linked_entries.get_text().split())
        print(linked_entriesF)

        
    print('\n')
    
    
```
